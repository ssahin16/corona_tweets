7- Replace all users names in the 'text' field with the usrusr token. You need to write a regular expression that identify user names and use the replace method
   to replace it with the 'usrusr' token. For instance:

"Are you there @ahurriyetoglu" should be "Are you there usrusr"

```python
import pandas as pd
import re
ct['text'] = ct['text'].replace(
    to_replace='@\w+', value='usrusr', regex=True)
ct
```

```python
ct['text'] = ct['text'].replace(
    to_replace = 'http[^\ ]*', value='urlurl', regex=True)
ct['text']
```

```python
import collections
counter = collections.Counter()
for row in ct['extracted']:
    for word in row:
        counter[word] += 1

for word in counter.most_common(100):
    print(word[0])
```
